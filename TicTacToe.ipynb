**Creating Tic Tac Toe Using OOP**

```python
import random


class TicTacToe:

    def __init__(self):
        self.board = []

    def display_board(self):
        print (self.board[0],'l',self.board[1],'l',self.board[2])
        print('------------')
        print (self.board[3],'l',self.board[4],'l',self.board[5])
        print('------------')
        print (self.board[6],'l',self.board[7],'l',self.board[8])
        print('------------')

    def create_board(self):
        self.board=list(range(0,9))
        self.display_board()
        self.board = [' ' , ' ', ' ',' ' , ' ' , ' ' , ' ' , ' ', ' ']


    def get_random_first_player(self):
        return random.randint(1, 2)

    def place_marker(self, marker, position):
        if self.board[position] == ' ':
           self.board[position] = marker

        else:
          print("That place is already filled.\n Move to which place?")
          position = int(input("Please pick a position from 0 to 8 again"))
          self.board[position] = marker

    def choose_marker(self):
          marker = input("Please pick a marker 'X' or 'O'.....")
          return marker

    def is_player_win(self, player):
        if self.board[0] == self.board[1] == self.board[2] != ' ': # 1st row
            return True
        elif self.board[3] == self.board[4] == self.board[5] != ' ': # 2nd row
            return True
        elif self.board[6] == self.board[7] == self.board[8] != ' ': # 3rd row
            return True
        elif self.board[0] == self.board[3] == self.board[6] != ' ': # 1st column
            return True
        elif self.board[1] == self.board[4] == self.board[7] != ' ': # 2nd column
            return True
        elif self.board[2] == self.board[5] == self.board[8] != ' ': # 3rd column
            return True
        elif self.board[0] == self.board[4] == self.board[8] != ' ': # diagonal \
            return True
        elif self.board[2] == self.board[4] == self.board[6] != ' ': # diagonal /
            return True
        else:
            return False


    def is_board_filled(self):
        for row in self.board:
            for i in row:
                if i == ' ':
                    return False
        return True

    def swap_player_turn(self, player):
        return 'Player1' if player == 'Player2' else 'Player2'

    def swap_marker_turn(self, marker):
        return 'X' if marker == 'O' else 'O'


    def start(self):
        print('Welcome to Tic Tac Toe!')
        self.create_board()
        # Who Starts First
        print('Randomly decide which player goes first....')
        player = 'Player1' if self.get_random_first_player() == 1 else 'Player2'
        print(f" {player} will go first.")

        marker = 'X' if self.choose_marker() == 'X' else 'O'

        while True:
            self.display_board()

            # taking user input

            position = int(input("Please pick a position from 0 to 8....."))
            self.place_marker( marker, position)


            # checking whether current player is won or not
            if self.is_player_win(player):
                print()
                print(f"Congratulations! {player} wins the game!")
                break

            # checking whether the game is draw or not
            if self.is_board_filled():
                print()
                print("\nGame Over.\n")
                print('The game is a draw!')
                break

            # swapping the turn
            player = self.swap_player_turn(player)
            print(f" {player} turn.")
            # swapping the Marker
            marker = self.swap_marker_turn(marker)

        # showing the final view of board
        print()
        self.display_board()


# starting the game
tic_tac_toe = TicTacToe()
tic_tac_toe.start()

restart = input("Do want to Replay Again?(y/n)......")
if restart == "n" or restart == "N":
    print('***See you next time....!! :(**** ')
else:
  tic_tac_toe = TicTacToe()
  tic_tac_toe.start()
```
